```python
aggregations={
    'Fare':{
        'median_Fare':'median',
        'mean_Fare':'mean',
        'max_Fare':max,
        'min_Fare':min
    },
    'Age':{
        'median_Age':'median',
        'min_Age':min,
        'max_Age':max,
        'range':lambda x:max(x)-min(x)
    }
}

df.groupby('Pclass').agg(aggregations)
```
